Check whether a word is bengali


```python
import unicodedata
import string
def is_bangla(word):
    try:
        lang = unicodedata.name(word.strip()[0])
        if 'BENGALI' in lang:
            return True
        else:
            return False
    except:
        return False
```

# Preprocessing

```python
import string
```

```python
string.punctuation
```

```python
punctuations='!#$%&\'()*+,-./:;<=>?@[\\]^_`{}~'
```

```python
bengali_number = ['০', '১','২', '৩','৪','৫','৬', '৭', '৮', '৯']
```

```python
def is_number_bengali(num):
    for digit in num:
        if digit in bengali_number:
            return True
    return False
```

```python
def clean_corpus(corpus,file_handle):
    for i in range(0,len(corpus)):
        item = corpus[i]
        transtable =  str.maketrans(punctuations, ' '*len(punctuations)) #map punctuation to space
        item = item.translate(transtable)
        item=" ".join(item.split("\u200e"))
        item=" ".join(item.split("\u200c"))
        item=" ".join(item.split("\u200d"))
        ta_sentence = [word for word in item.split() if (is_bangla(word))]
        sentence=""
        for word in ta_sentence:
            if not is_number_bengali(word):
                #file_handle.write('%s\n' % word)
                sentence = sentence+" "+word
        file_handle.write('%s\n' % sentence)
        sentence = ""    
```

```python
def clean_words(corpus,file_handle):
    for i in range(0,len(corpus)):
        item = corpus[i]
        transtable =  str.maketrans(punctuations, ' '*len(punctuations)) #map punctuation to space
        item = item.translate(transtable)
        ta_sentence = [word for word in item.split() if (is_bangla(word))]
        for word in ta_sentence:
            if not is_number_bengali(word):
                file_handle.write('%s\n' % word)
```

#Extracting the vector representation of words from HAL

```python
def extract_vector_from_l2r(word):
  word_vector=[]
  for i in sorted_vocab_dict.keys():
    try:
      word_vector.append(hal_l2r[vocab_index[word],vocab_index[i]])
    except KeyError:
      word_vector.append(0)
  return word_vector  

def extract_vector_from_r2l(word):
  word_vector=[]
  for i in sorted_vocab_dict.keys():
    try:
      word_vector.append(hal_l2r[(vocab_index[i],vocab_index[word])])
    except KeyError:
      word_vector.append(0)
  return word_vector  
```

# Combining Word vectors

```python
def combined_word_vector(word):
    l2r_vec=extract_vector_from_l2r(word)
    r2l_vec=extract_vector_from_r2l(word)
    return([(x + y)/2 for x, y in zip(l2r_vec, r2l_vec)])
```
